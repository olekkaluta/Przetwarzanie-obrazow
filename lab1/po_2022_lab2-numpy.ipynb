(**0.4 punktu**) Przeprowadź <u>symulację</u> $n = 10000$ eksperymentów z wielokrotnym rzutem sześcienną kostką do gry. W każdym $i$-tym eksperymencie zasymuluj $i \cdot 10$-krotny rzut kostką. Przykład: w pierwszym eksperymencie symulujesz  10 rzutów, w drugim 20 rzutów, w setnym 1000 rzutów, a w dziesięciotysięcznym 100000 rzutów.
    
Każdy pojedynczy rzut kostką to nic innego jak generowanie całkowitej liczby losowej z przedziału $1..6$ (czyli: ``np.random.randint(1,7, i)``).

W każdym $i$-tym eksperymencie wyznacz częstość wyrzucenia określonej liczby oczek, np. "1". Ty wybierasz, która liczba oczek interesuje Cię w całym eksperymencie. W każdym rzucie każdego eksperymentu interesuje Cię tylko ta jedna ustalona liczba oczek.

Przedstaw na wykresie krzywą reprezentującą częstość wyrzuceń interesującej Cię liczby oczek (ścianki kostki; oś $y$ wykresu) w funkcji numeru eksperymentu (oś $x$ wykresu). 
Jakie jest prawdopodobieństwo wyrzucenia dowolnej określonej liczby oczek (np. "1") w serii nieskończenie wielu rzutów kostką?
Sprawdź czy Twoja krzywa stabilizuje się na tej wartości teoretycznej.

Twój wynik powinien być następującej postaci:



*tu umieść swoje rozwiązanie*

```python
import numpy as np
import matplotlib.pyplot as plt
n = 10000+1
cyfra = 6
tablica_trafien = []
for i in range(1,n):
    rzuty = i * 10
    wynik = np.random.randint(1,7, rzuty)
    ilosc_trafien = (np.sum(wynik==cyfra) / wynik.size)
    tablica_trafien.append(ilosc_trafien)

plt.plot(tablica_trafien)
plt.ylabel('ilosc trafien')
plt.axhline(y=1/6,color='r',linestyle='--')
plt.show()


```
