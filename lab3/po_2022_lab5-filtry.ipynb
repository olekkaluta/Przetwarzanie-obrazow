```python
import numpy as np
import matplotlib.pyplot as plt
import copy

img = plt.imread('butterfly.jpg')
plt.imshow(img)

r=img[:,:,0]
g=img[:,:,1]
b=img[:,:,2]
gray=0.299*r+0.587*g+0.114*b
plt.imshow(gray)

img2 = gray.copy()
a=[]
d=[]


for x in range(img2.shape[0]-1):
    for y in range(img2.shape[1]-1):
        for dx in range(-1, 2):
            for dy in range(-1, 2):
               a.append(img2[x+dx][y+dy])    

c=np.reshape(a,((img2.shape[0]-1),(img2.shape[1]-1),9))
sr=np.mean(c, axis=2)
m=np.median(c, axis=2)

for x in range(img2.shape[0]-5):
    for y in range(img2.shape[1]-5):
        for dx in range(-5, 6):
            for dy in range(-5, 6):
               d.append(img2[x+dx][y+dy])

c2=np.reshape(d,((img2.shape[0]-5),(img2.shape[1]-5),121))
sr2=np.mean(c2, axis=2)
m2=np.median(c2, axis=2)
```
